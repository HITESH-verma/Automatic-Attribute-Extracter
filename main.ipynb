```python
import pandas as pd
import os
import cv2
import pytesseract
import re
from concurrent.futures import ThreadPoolExecutor
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Function to extract text and search for values matching the specific label and its units
def search_label_values(image_path, label, config=r"--psm 11 --oem 1"):
    try:
        # Read the image using OpenCV
        img = cv2.imread(image_path)
        
        if img is None:
            return f"Image not found or unable to open: {image_path}"

        # Convert to grayscale
        gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        
        # Apply adaptive thresholding
        thresh_img = cv2.adaptiveThreshold(
            gray_img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2
        )
        
        # Extract text from the thresholded image using pytesseract
        text = pytesseract.image_to_string(thresh_img, config=config)
        
        # Replace newlines with spaces for easier searching
        text_clean = text.replace('\n', ' ')
        

        if label == "item_weight":
            # Create a regex pattern for capturing numbers followed by specific units
            unit_pattern = r'\d+\.?\d*\s?(?:g|kg|mg|milligram|gram|kilogram|ounce|ton|pound)'
        elif label == "width" or label == "depth" or label == "height":
            unit_pattern = r'\d+\.?\d*\s?(?:cm|mm|m|centimetre|millimetre|metre|inch|foot|yard)'
        elif label == "voltage":
            unit_pattern = r'\d+\.?\d*\s?(?:v|mv|kv|volt|millivolt|kilovolt)'
        elif label == "wattage":
            unit_pattern = r'\d+\.?\d*\s?(?:w|kw|watt|kilowatt)'
        elif label == "item_volume":
            unit_pattern = r'\d+\.?\d*\s?(?:l|ml|litre|millilitre|cubic foot|gallon|cup|fluid ounce|pint|quart)'
        else:
            return f"Label '{label}' not recognized"
        
        # Find all matches in the extracted text (case insensitive)
        matches = re.findall(unit_pattern, text_clean, re.IGNORECASE)

        return matches if matches else f"No values found for label '{label}'"
    
    except Exception as e:
        return f"An error occurred: {e}"

# Process each image in parallel using ThreadPoolExecutor
def process_row(i, row, image_folder):
    label = row.iloc[1]  # Assuming the second column contains the label
    image_filename = f'image_{i}.jpg'  # Use i for sequential image names
    image_path = os.path.join(image_folder, image_filename)

    if os.path.exists(image_path):
        return search_label_values(image_path, label)
    else:
        return 'Image missing'

def process_images_and_update_csv(image_folder, csv_file_path, output_csv_path):
    # Load the CSV
    df = pd.read_csv(csv_file_path)

    # Ensure the 'found_value' column exists in the dataframe
    if 'found_value' not in df.columns:
        df['found_value'] = ''  # Initialize the column

    # Process each row in parallel
    with ThreadPoolExecutor() as executor:
        results = executor.map(lambda i: process_row(i, df.iloc[i], image_folder), range(1, 100))

    # Update the CSV with the results
    for i, result in enumerate(results, 1):  # Start from index 1 to match the row numbering
        df.at[i, 'found_value'] = result

    # Save the updated CSV to a new file
    df.to_csv(output_csv_path, index=False)
    print(f"CSV file updated and saved to {output_csv_path}")

# Example usage
image_folder = r'C:\Users\Hitesh\Desktop\amazon ml\train images'
csv_file_path = r'C:\Users\Hitesh\Desktop\amazon ml\train(3000).csv'
output_csv_path = r'C:\Users\Hitesh\Desktop\amazon ml\updated_train_1.csv'

process_images_and_update_csv(image_folder, csv_file_path, output_csv_path)
```

```python
import pandas as pd
import os
import cv2
import numpy as np
from paddleocr import PaddleOCR
from concurrent.futures import ThreadPoolExecutor

# Initialize PaddleOCR
ocr = PaddleOCR(use_angle_cls=True, lang='en')

# Function to apply grayscale and thresholding
def preprocess_image(image_path):
    # Read the image using OpenCV
    img = cv2.imread(image_path)
    
    if img is None:
        raise FileNotFoundError(f"Image not found or unable to open: {image_path}")
    
    # Convert to grayscale
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Apply adaptive thresholding
    thresh_img = cv2.adaptiveThreshold(
        gray_img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2
    )
    
    return thresh_img

# Function to extract text and search for values matching the specific label and its units
def search_label_values_paddleocr(image_path, label):
    try:
        # Preprocess the image
        preprocessed_image = preprocess_image(image_path)
        
        # Save the preprocessed image to a temporary path
        temp_path = image_path.replace('.jpg', '_temp.jpg')
        cv2.imwrite(temp_path, preprocessed_image)
        
        # Perform OCR on the preprocessed image
        ocr_result = ocr.ocr(temp_path, cls=True)
        
        # Extract text from the OCR result
        text_clean = ' '.join([line[1][0] for line in ocr_result[0]])
        
        # Remove the temporary preprocessed image file
        os.remove(temp_path)

        # Define the regular expression patterns for each label
        if label == "item_weight":
            unit_pattern = r'\d+\.?\d*\s?(?:g|kg|mg|milligram|gram|kilogram|ounce|ton|pound)'
        elif label in ["width", "depth", "height"]:
            unit_pattern = r'\d+\.?\d*\s?(?:cm|mm|m|centimetre|millimetre|metre|inch|foot|yard)'
        elif label == "voltage":
            unit_pattern = r'\d+\.?\d*\s?(?:v|mv|kv|volt|millivolt|kilovolt)'
        elif label == "wattage":
            unit_pattern = r'\d+\.?\d*\s?(?:w|kw|watt|kilowatt)'
        elif label == "item_volume":
            unit_pattern = r'\d+\.?\d*\s?(?:l|ml|litre|millilitre|cubic foot|gallon|cup|fluid ounce|pint|quart)'
        else:
            return f"Label '{label}' not recognized"

        # Search for matches using the regex pattern
        matches = re.findall(unit_pattern, text_clean, re.IGNORECASE)

        return matches if matches else f"No values found for label '{label}'"

    except Exception as e:
        return f"An error occurred: {e}"

# Function to process each row (for multithreading)
def process_row(i, row, image_folder):
    label = row.iloc[1]  # Assuming the second column contains the label
    image_filename = f'image_{i}.jpg'  # Use i for sequential image names
    image_path = os.path.join(image_folder, image_filename)

    if os.path.exists(image_path):
        return search_label_values_paddleocr(image_path, label)
    else:
        return 'Image missing'

# Function to process images and update the CSV file
def process_images_and_update_csv(image_folder, csv_file_path, output_csv_path):
    # Load the CSV
    df = pd.read_csv(csv_file_path)

    # Ensure the 'found_value' column exists in the dataframe
    if 'found_value' not in df.columns:
        df['found_value'] = ''  # Initialize the column

    # Process each row in parallel using ThreadPoolExecutor
    with ThreadPoolExecutor() as executor:
        results = executor.map(lambda i: process_row(i, df.iloc[i], image_folder), range(1, 101))  # Process first 100 images

    # Update the CSV with the results
    for i, result in enumerate(results, 1):  # Start from index 1 to match the row numbering
        df.at[i, 'found_value'] = result

    # Save the updated CSV to a new file
    df.to_csv(output_csv_path, index=False)
    print(f"CSV file updated and saved to {output_csv_path}")

# Example usage
image_folder = r'C:\Users\Hitesh\Desktop\amazon ml\train images'
csv_file_path = r'C:\Users\Hitesh\Desktop\amazon ml\train(3000).csv'
output_csv_path = r'C:\Users\Hitesh\Desktop\amazon ml\updated_train_paddle1.csv'

process_images_and_update_csv(image_folder, csv_file_path, output_csv_path)
```

```python
import pandas as pd
import os
import cv2
import numpy as np
from paddleocr import PaddleOCR
import re

# Initialize PaddleOCR
ocr = PaddleOCR(use_angle_cls=True, lang='en')

# Folder containing images
image_folder = r'C:\Users\Hitesh\Desktop\amazon ml\train2\\'

# CSV file path where ID and match value will be saved
output_csv = r'C:\Users\Hitesh\Desktop\amazon ml\updated_train_paddle.csv'

# Number of images to process
num_images_to_process = 200

# Load the original CSV with image IDs and labels
input_csv = 'train(3000).csv'  # Assuming this is your file with 'group_id' and 'entity_name' columns
df = pd.read_csv(input_csv)

# Ensure you process only the first `num_images_to_process` rows
df = df.head(num_images_to_process)

# To store the results
results = []

# Loop over each row in the dataframe and process the image
for index, row in df.iterrows():
    group_id = row['group_id']  # Assuming your input CSV has a 'group_id' column
    label = row['entity_name']  # Assuming your input CSV has an 'entity_name' column
    
    # Construct image file path with sequential numbering (image_1.jpg, image_2.jpg, etc.)
    image_path = os.path.join(image_folder, f"image_{index + 1}.jpg")  # Start from "image_1.jpg"
    
    img = cv2.imread(image_path)
    
    if img is None:
        print(f"Image not found or unable to open: {image_path}")
        # Append empty value if image is not found
        results.append({'group_id': group_id, 'match_value': ''})
        continue
    
    try:
        # Perform OCR on the image
        ocr_result = ocr.ocr(image_path, cls=True)
        
        # Check if OCR result contains any data
        if not ocr_result or not ocr_result[0]:
            print(f"No OCR results found for image: {image_path}")
            results.append({'group_id': group_id, 'match_value': ''})
            continue
        
        # Extract text from the OCR result
        text_clean = ' '.join([line[1][0] for line in ocr_result[0]])

        # Define the regular expression patterns for each label
        if label == "item_weight":
            unit_pattern = r'\d+\.?\d*\s?(?:g|kg|mg|milligram|gram|kilogram|ounce|ton|pound)'
        elif label in ["width", "depth", "height"]:
            unit_pattern = r'\d+\.?\d*\s?(?:cm|mm|m|centimetre|millimetre|metre|inch|foot|yard)'
        elif label == "voltage":
            unit_pattern = r'\d+\.?\d*\s?(?:v|mv|kv|volt|millivolt|kilovolt)'
        elif label == "wattage":
            unit_pattern = r'\d+\.?\d*\s?(?:w|kw|watt|kilowatt)'
        else:
            unit_pattern = r'\d+\.?\d*\s?(?:l|ml|litre|millilitre|cubic foot|gallon|cup|fluid ounce|pint|quart)'
        
        # Find all matches using regex
        matches = re.findall(unit_pattern, text_clean, re.IGNORECASE)
        
        # Save the first match or leave empty if no match is found
        match_value = matches[0] if matches else ''
    
    except Exception as e:
        print(f"Error processing image {image_path}: {e}")
        match_value = ''  # Leave empty in case of any error
    
    # Append the result to the results list
    results.append({'group_id': group_id, 'match_value': match_value})

# Convert the results to a DataFrame
output_df = pd.DataFrame(results)

# Save the results to a new CSV file
output_df.to_csv(output_csv, index=False)

print(f"Results saved to {output_csv}")
```

```python
import pandas as pd
import re

# Function to normalize numeric values for comparison
def normalize_value(value):
    if pd.isna(value) or value == '':  # Handle NaN or empty values
        return None
    try:
        # Remove any non-numeric characters, including spaces and units (if any)
        # This ensures that '008 g' and '8.0' are treated as '8.0'
        value = re.sub(r'[^\d\.]', '', str(value)).strip()  # Remove all non-numeric characters
        return float(value)  # Convert to float
    except ValueError:
        return None

# Load the CSV file
input_csv = 'updated_train_paddle1.csv'  # Replace with your file path
df = pd.read_csv(input_csv)

# Create the 'label' column to store whether values are equal
df['label'] = 0  # Initialize as unequal (0)

# Compare the 'predicted' and 'original' columns
for index, row in df.iterrows():
    predicted_value = normalize_value(row['predicted'])
    original_value = normalize_value(row['original'])

    # Check if both values are not None and if they're close enough to be considered equal
    if predicted_value is not None and original_value is not None:
        if abs(predicted_value - original_value) < 1e-3:  # Allow small tolerance
            df.at[index, 'label'] = 1  # Mark as equal (1)

# Save the updated CSV file with the new 'label' column
output_csv = 'updated_file.csv'
df.to_csv(output_csv, index=False)

print(f"Updated CSV with 'label' column saved as {output_csv}.")
df['label1'] = 1
true_labels = df['label1']
from sklearn.metrics import f1_score
# Predicted labels are also the 'label' column after the comparison
predicted_labels = df['label']  # This would ideally be different if you had an external prediction

# Calculate F1 score
f1 = f1_score(true_labels, predicted_labels)

print(f"F1 Score: {f1}")
```
